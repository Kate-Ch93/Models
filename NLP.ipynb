```python
from textblob import TextBlob

# Пример текста
text = "I love natural language processing!"

# Создаём объект TextBlob
blob = TextBlob(text)

# Токенизация на слова
words = blob.words
print("Words:", words)

# Анализ сентимента
sentiment = blob.sentiment
print("Sentiment:", sentiment)

# Лемматизация
lemmatized = blob.words.lemmatize()
print("Lemmatized:", lemmatized)
```

```python
#pipeline
import spacy
#import numpy

# Загружаем модель для русского языка (нужно установить один раз через команду в терминале)
# pip install spacy
# python -m spacy download ru_core_news_sm

nlp = spacy.load("ru_core_news_sm") # предобученная модель, делает токенизацию, лемматизацию, удаляет пунктуацию, можно получать части речи

text = 'Привет! Как дела? Что делаешь сегодня?'

# Обработка текста
doc = nlp(text.lower()) #1

# Извлечение лемм, фильтрация только алфавитных токенов
lemmas = [token.lemma_ for token in doc if token.is_alpha] # 2,4,5 (токенизация, лемматизация, удаление пунктуации и символов)

# Оставляем только те слова, которые НЕ являются стоп-словами и НЕ знаками препинания
filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]  # 3 удаление стоп-слов

print("Леммы:", lemmas)
print(filtered_tokens)
```

```python
#pipeline для очистки
import re
import pymorphy2
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

morph = pymorphy2.MorphAnalyzer()
stop_words = set(stopwords.words('russian'))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text, language='russian')
    tokens = [word for word in tokens if word not in stop_words]
    lemmas = [morph.parse(word)[0].normal_form for word in tokens]
    return ' '.join(lemmas)
```
